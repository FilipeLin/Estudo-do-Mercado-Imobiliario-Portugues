```python
import pandas as pd

df=pd.read_csv('portugal_listinigs.csv',low_memory=False)

df.info()
```

```python
# Excluindo as colunas especificadas
colunas_para_apagar = ['GrossArea', 'HasParking', 'Floor', 'EnergyEfficiencyLevel', 
                       'ElectricCarsCharging', 'NumberOfWC', 'ConservationStatus', 
                       'LotSize', 'BuiltArea']

# Remover as colunas
df = df.drop(columns=colunas_para_apagar)

# Exibindo as primeiras linhas para verificar se as colunas foram removidas
print(df.head())
```

```python
# Definindo as colunas que você deseja alterar os valores nulos
colunas_para_preencher = ['Town', 'Type', 'EnergyCertificate', 'Parking', 
                          'ConstructionYear', 'Garage', 'Elevator', 
                          'TotalRooms', 'NumberOfBedrooms', 'NumberOfBathrooms']

# Preenchendo os valores nulos nas colunas com a string 'Unknown'
df[colunas_para_preencher] = df[colunas_para_preencher].fillna('Unknown')

# Exibindo as primeiras linhas para verificar
print(df.head())
```

```python
df.info()
```

```python
# Agrupar os dados pelas colunas "Type" e calcular a media de "TotalArea"
media_totalarea = df.groupby(['Type'])['TotalArea'].mean()
 
# Função para preencher os valores nulos com a média calculada
def preencher_totalarea(row):
    if pd.isnull(row['TotalArea']):
        # Tenta obter a média com base nas 4 colunas
        return media_totalarea.get((row['Type']))
    return row['TotalArea']
 
# Aplicando a função para preencher os valores nulos
df['TotalArea'] = df.apply(preencher_totalarea, axis=1)
 
# Exibindo as primeiras linhas para verificar
print(df.head())
```

```python
# Agrupar os dados pelas colunas "Type" e calcular a media de "LivingArea"
media_livingarea = df.groupby(['Type'])['LivingArea'].mean()
 
# Função para preencher os valores nulos com a média calculada
def preencher_livingarea(row):
    if pd.isnull(row['LivingArea']):
        # Tenta obter a média com base nas 4 colunas
        return media_livingarea.get((row['Type']))
    return row['LivingArea']
 
# Aplicando a função para preencher os valores nulos
df['LivingArea'] = df.apply(preencher_livingarea, axis=1)
 
# Exibindo as primeiras linhas para verificar
print(df.head())
```

```python
# Agrupar os dados pelas colunas "Type", "District", "City", "Town" e "TotalArea" e calcular a média de "Price"
media_preco = df.groupby(['Type', 'District', 'City', 'Town', 'TotalArea'])['Price'].mean()
 
# Função para preencher os valores nulos com a média calculada
def preencher_preco(row):
    if pd.isnull(row['Price']):
        # Tenta obter a média com base nas 4 colunas
        return media_preco.get((row['Type'], row['District'], row['City'], row['Town'], row['TotalArea']), row['Price'])
    return row['Price']
 
# Aplicando a função para preencher os valores nulos
df['Price'] = df.apply(preencher_preco, axis=1)
 
# Exibindo as primeiras linhas para verificar
print(df.head())
```

```python
df.isna().sum()
```

```python
# Remover as linhas que ainda têm valores nulos na coluna 'Price'
df = df.dropna(subset=['Price'])
```

```python
df.to_csv('portugal_listinigs_modificado.csv', index=False)
```
